#1. Activation Functions

Activation functions introduce non-linearity to neural networks so they can learn complex patterns.


## 1.Sigmoid

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

Range: (0, 1)

**Pros:**

*   Good for probabilities (output between 0 and 1)
*   Historically used in early neural nets

**Cons (Vanishing Gradient Problem):**

*   For very large or very small x, gradient ‚Üí 0


*   This kills learning, especially in deep networks

*   Not used in hidden layers anymore


```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Test
for val in [-5, -1, 0, 1, 5]:
    print(val, sigmoid(val), sigmoid_derivative(val))
```

#2. Tanh

tanh(x) = (e^x ‚àí e^(‚àíx)) / (e^x + e^(‚àíx))

It outputs values between -1 and +1.

Sigmoid outputs between 0 and 1

This causes neurons to be always positive, which shifts data and slows learning.

But tanh is zero-centered:

Large positive x ‚Üí output ‚âà +1

Large negative x ‚Üí output ‚âà -1

x = 0 ‚Üí output = 0

This is better for hidden layers because gradients flow more symmetrically.

```python
import math
def tanh(x): return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
def tanh_deriv_direct(x): return 4 / (math.exp(x) + math.exp(-x))**2
def tanh_deriv_identity(x): s = tanh(x); return 1 - s*s

x = 1.0
print(tanh(x), tanh_deriv_direct(x), tanh_deriv_identity(x))
```

# Ridge Regression

```python
import numpy as np
import pandas as pd
```

# Data

Feature	Meaning

x1	Study Hours

x2	Attendance

x3	Previous Score

x4	Random Noise

x5	Random Noise

y	Final Score

```python
X = [
    [2, 80, 70, 15, 200],
    [3, 85, 72, 18, 180],
    [5, 90, 78, 14, 210],
    [6, 92, 85, 20, 190],
    [8, 95, 88, 16, 220],
    [9, 96, 91, 21, 230],
    [11, 98, 94, 19, 240],
    [12, 99, 96, 17, 250],
    [14, 100, 98, 22, 260],
    [15, 100, 99, 18, 270]
]

X = np.array(X)

y = [
    65,
    68,
    75,
    80,
    86,
    90,
    95,
    97,
    99,
    100
]

y = np.array(y)
```

# Normalization

```python
X = (X - X.mean(axis=0)) / X.std(axis=0)
X
```

# W (Weight) and b (bais)

```python
W = np.full(X.shape[1], 0.5)
b = 0.5
lam = 0.1
```

# Predict

```python
def predict(x, W, b):
    return np.dot(x, W) + b

predict(X, W, b)
```

# Loss

```python
def lasso_MSE(x, y, W, b, lam):

    # Lam controll the loss

    n = x.shape[0]
    MSE_loss = np.mean((y - predict(x, W, b))**2)
    lasso_loss = MSE_loss + (lam * np.sum(np.abs(W)))
    return lasso_loss

lasso_MSE(X, y, W, b, lam)
```

# Learning or gradient Descent

```python
def gradient_descent(X, y, W, b, lam):
    n = X.shape[0]

    dw = (-2/n * np.dot(X.T, (y - predict(X, W, b)))) + (lam * np.sign(W))
    db = (-2/n * np.sum(y - predict(X, W, b)))

    # print(dw)
    # print(db)
    lr = 0.001
    W = W - (lr * dw)
    b = b - (lr * db)

    return W, b

# gradient_descent(X, y, W, b, 2)
```

# Iterate for 50 epocs

```python
epochs = 10000
for epoch in range(epochs):
    loss = lasso_MSE(X, y, W, b, lam)
    W, b = gradient_descent(X, y, W, b, lam)
    if epoch % 1000 == 0:
        print("predict : ", predict(X, W, b))
        print("Loss : ", loss)
        print("W : ", W)
        print("b : ", b)
```

# Elastic net Regression

```python
import numpy as np
```

# Dataset

```python
X = [
    [2, 80, 70, 15, 200],
    [3, 85, 72, 18, 180],
    [5, 90, 78, 14, 210],
    [6, 92, 85, 20, 190],
    [8, 95, 88, 16, 220],
    [9, 96, 91, 21, 230],
    [11, 98, 94, 19, 240],
    [12, 99, 96, 17, 250],
    [14, 100, 98, 22, 260],
    [15, 100, 99, 18, 270]
]

X = np.array(X)

y = [
    65,
    68,
    75,
    80,
    86,
    90,
    95,
    97,
    99,
    100
]

y = np.array(y)
```

# Normalization

```python
X = (X - X.mean(axis=0)) / X.std(axis=0)
X
```

# Parameters

```python
W = np.full(X.shape[1], 0.5)
b = 0.5
l1 = 0.01
l2 = 0.01
lr = 0.001
```

# Prediction

```python
def predict(X, W, b):
    return np.dot(X, W) + b

predict(X, W, b)
```

# Loss

```python
def elastic_MSE(X, y, W, b, l1, l2):
    mse_loss = np.mean((y-predict(X, W, b))**2)
    ridge_loss = l2 * np.sum(W**2)
    lasso_loss = l1 * np.sum(np.abs(W))
    
    elastic_loss = mse_loss + ridge_loss + lasso_loss

    # print(ridge_loss)
    # print(lasso_loss)
    # print(mse_loss)
    
    return elastic_loss

elastic_MSE(X, y, W, b, l1, l2)
```

# Gradient descent

```python
def gradient_descent(X, y, W, b, l1, l2, lr):
    
    n = X.shape[0]

    dW = ((-2/n) * np.dot(X.T, (y - predict(X, W, b)))) + (2 * l2 * W) + (l1 * np.sign(W))
    db = (-2/n) * np.sum(y - predict(X, W, b))

    W = W - (dW * lr)
    b = b - (db * lr)

    return W, b
```

```python
epochs = 10000
for epoch in range(epochs):
    loss = elastic_MSE(X, y, W, b, l1, l2)
    W, b = gradient_descent(X, y, W, b, l1, l2, lr)
    if epoch % 1000 == 0:
        print("predict : ", predict(X, W, b))
        print("Loss : ", loss)
        print("W : ", W)
        print("b : ", b)
```

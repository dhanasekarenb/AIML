# Ridge Regression

```python
import numpy as np
import pandas as pd
```

# Data

```python
X = np.array([
    [2, 7, 80, 5],
    [3, 6, 85, 4],
    [4, 8, 90, 3],
    [5, 5, 70, 6],
    [6, 7, 95, 2],
    [7, 6, 88, 2],
    [8, 8, 98, 1],
    [2, 5, 60, 7],
    [9, 7, 99, 1],
    [1, 4, 50, 8]
])

y = np.array([
    40,
    50,
    65,
    55,
    75,
    80,
    92,
    35,
    96,
    25
])
```

# Normalization

```python
X = (X - X.mean(axis=0)) / X.std(axis=0)
```

# W (Weight) and b (bais)

```python
W = np.full(4, 0.5)
b = 0.5
lam = 0.1
```

# Predict

```python
def predict(x, W, b):
    return np.dot(x, W) + b

predict(X, W, b)
```

# Loss

```python
def ridge_MSE(x, y, W, b, lam):

    # Lam controll the loss

    n = x.shape[0]
    MSE_loss = np.mean((y - predict(x, W, b))**2)
    ridge_loss = MSE_loss + (lam * np.sum(W ** 2))
    return ridge_loss

ridge_MSE(X, y, W, b, lam)
```

# Learning or gradient Descent

```python
def gradient_descent(X, y, W, b, lam):
    n = X.shape[0]

    dw = (-2/n * np.dot(X.T, (y - predict(X, W, b)))) + (lam * 2 * W)
    db = (-2/n * np.sum(y - predict(X, W, b)))

    # print(dw)
    # print(db)
    lr = 0.001
    W = W - (lr * dw)
    b = b - (lr * db)

    return W, b

# gradient_descent(X, y, W, b, 2)
```

# Iterate for 50 epocs

```python
epochs = 10000
for epoch in range(epochs):
    loss = ridge_MSE(X, y, W, b, lam)
    W, b = gradient_descent(X, y, W, b, lam)
    if epoch % 1000 == 0:
        print("predict : ", predict(X, W, b))
        print("Loss : ", loss)
        print("W : ", W)
        print("b : ", b)
```

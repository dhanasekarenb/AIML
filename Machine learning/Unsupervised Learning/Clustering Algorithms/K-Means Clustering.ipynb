# K-Means Clustering

```python
import numpy as np
```

# Data

```python
X = np.array([
    [1, 2],
    [2, 3],
    [3, 1],
    [8, 8],
    [9, 10],
    [10, 9]
])
```

# Parameter

```python
# Number of Clusters
K = 2

# Initial Centroids
C = {
    0: [1, 2],
    1: [8, 8]
}
```

# Calculate Distance

```python
def distance(point, centroid):
    return np.sqrt(np.sum((np.array(point) - np.array(centroid)) ** 2))
```

# Assign Points to Nearest Centroid

```python
def min_dist(X, C):

    cluster = {}

    for i in range(K):
        cluster[i] = []

    for x in X:

        distance_data = []

        for i in range(K):

            d = distance(x, C[i])

            distance_data.append(d)

        nearest_centroid = distance_data.index(min(distance_data))

        cluster[nearest_centroid].append(x.tolist())

    return cluster
```

```python
# Recalculate Centroids
def recalc_dist(cluster, C):

    for i in range(K):

        points = np.array(cluster[i]) # cluster 0 is 
        centroid = np.mean(points, axis=0)
        C[i] = centroid.tolist()
```

# Training Loop

```python
# Training Loop
cluster = {}

for epoch in range(100):

    temp_cluster = min_dist(X, C)

    # Stop if clusters do not change
    if cluster == temp_cluster:

        print(f"Converged at epoch : {epoch}")
        break

    cluster = temp_cluster

    recalc_dist(cluster, C)
```

# Final Output

```python
# Final Output
print("\nFinal Clusters:\n")

for i in cluster:

    print(f"Cluster {i} : {cluster[i]}")


print("\nFinal Centroids:\n")

for i in C:

    print(f"C{i} : {C[i]}")
```
